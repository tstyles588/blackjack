```python
from random import shuffle
import random
import sys

ranks = ['Spades', 'Clubs', 'Hearts', "Diamonds"]
suits = list(range(1, 14))
deck = [(suit, rank) for rank in ranks for suit in suits]

class Deck:
    ranks = ['Spades', 'Clubs', 'Hearts', "Diamonds"]
    suits = list(range(1, 14))
    cards = [(suit, rank) for rank in ranks for suit in suits]

    
    def shuffle(self):
        random.shuffle(self.cards)
        
class Player:
    def __init__(self):
        self.hand = []
        self.total = 0
       
    def score_tally(self):
        self.total = 0
        for c in self.hand:
            self.total += c[0]
        return int(self.total)
    
    def show_cards(self, ply):
        if isinstance(self, Human):
            print(f"Player hand: {self.hand} Total: ")
            
        if isinstance(self, Dealer) and len(self.hand) < 3:
                  print(f"Dealer hand: {self.hand[0]} Total: ")
                  print(f"{self.hand[0][0]}")
        
        elif isinstance(self, Dealer) and len(self.hand) > 2:
                  print(f"Dealer hand: {self.hand} Total: ")

            
class Dealer(Player):
    def __init__(self):
        super().__init__()
        
    def deal(self, ply, deck):
        ply.hand.append(deck.cards.pop())
        
        ply.hand.append(deck.cards.pop())
        
    def hit(self, ply, deck):
        ply.hand.append(deck.cards.pop())
        
    def dbust(self):
        print("Dealer has bust!")
            
    def dwin(self):
            print("Dealer Wins!") 
            
class Human(Player):
    def __init__(self):
        self.hand = []
        self.total = 0
        
    def pbust(self):
        print("Player has bust!")
        
    def pwin(self):
            print("Player Wins!") 
class Game:
    def __init__(self):
        self.dealer = Dealer()
        self.player = Human()
        self.players = [self.dealer, self.player]
        
        def game_over(self):
            print('Thanks 4 playing!')
            
        def push(self):
            print('Player and dealer tied!')
            
        def blackjack(self):
            print('Player scored 21! Blackjack win!')
            
def main():
    deck = Deck()
    player = Human()
    dealer = Dealer()
    game = Game()
    
    print('Shuffling Cards!')
    deck.shuffle()
    
  ##  flag = False
   ## while not flag:
     #   dealer.deal(player, deck)
      #  if player.score_tally() == 21:
       #     game.blacjack()
        #    flag = True
        #elif player.score_tally() <= 21:
        #    flag= True
        #    
        #else:
         #   flag = False
    dealer.deal(player, deck)
    
    dealer.deal(dealer, deck)
    
    player.show_cards(player)
    
    print(player.score_tally())
    
    dealer.show_cards(dealer)
    
    playing =  True
    while playing:
        choice = input('Hit or Stand ???')
        
        if choice.lower() == 'hit':
            dealer.hit(player, deck)
            player.show_cards(player)
            
            print(player.score_tally())
            
            dealer.show_cards(dealer)
            
            if player.score_tally() > 21:
                player.pbust()
                dealer.dwin()
                choice = input('player Wins! Play again? Y/N').lower()
                if choice == 'y':
                    main()
                elif choice ==  'n':
                    break
            if player.score_tally() == 21:
                game.blackjack()
                
                choice = input('player Wins! Play again? Y/N').lower()
                if choice == 'y':
                    main()
                elif choice ==  'n':
                    break
        if choice.lower() == 'stand':
            dealer.hit(dealer, deck)
            
            player.show_cards(player)
            
            print(player.score_tally())
            
            dealer.show_cards(dealer)
            
            print(dealer.score_tally())
            
            limbo = True
            while limbo:
                if dealer.score_tally() < 17:
                    dealer.hit(dealer, deck) 
                    player.show_cards(player)
                    dealer.show_cards(dealer)
                    continue
                elif dealer.score_tally() > 21:
                    dealer.dbust()
                    player.pwin()
                    choice = input('Play again? Y/N').lower()
                if choice == 'y':
                    main()
                elif choice ==  'n':
                    break
                    
                else:
                    limbo = False
            if dealer.score_tally() == player.score_tally():
                game.push()
            if dealer.score_tally() < player.score_tally():
                player.pwin()
                choice = input('Play again? Y/N').lower()
                if choice == 'y':
                    main()
                elif choice ==  'n':
                    break
            if dealer.score_tally() > player.score_tally():
                dealer.dwin()
                
        if choice.lower() == 'stop':
            game.game_over()
            break
    player.show_cards(player)
    dealer.show_cards(dealer)
main()
```
